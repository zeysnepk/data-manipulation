```python
import seaborn as sns
import numpy as np
import pandas as pd
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
#0-1 transformation
```

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
df["new_sex"] = lbe.fit_transform(df.sex)
```

```python
df
```

```python
df["new2_sex"] = df["sex"].cat.codes
```

```python
df
```

```python
#1 and others(0) tranformation
df["new_day"] = np.where(df.day.str.contains("Sun"),1,0)
```

```python
df
```

```python
#multi-class transformation
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
lbe.fit_transform(df.day)
```

```python
df["new2_day"] = lbe.fit_transform(df.day)
df
```

```python
#one-hot transformation
df_one_hot = pd.get_dummies(df,columns=["sex",],dtype=int)
```

```python
#converting continuous variable to categorical variable
from sklearn.preprocessing import KBinsDiscretizer
df = tips.copy()
dff = df.select_dtypes(include=["float64", "int64"])
dff.head()
```

```python
est = KBinsDiscretizer(n_bins=[3,2,2],encode="ordinal",strategy="quantile").fit(dff)
```
