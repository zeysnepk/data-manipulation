# ðŸ§½ Cleaning NaN Values

#### Â» Create a dataframe containing NaN values 

```python
import numpy as np
import pandas as pd
vector1 = np.array([5,3,6,np.nan,7,2,8,np.nan,4,np.nan])
vector2 = np.array([10,1,5,3,2,8,np.nan,6,np.nan,np.nan])
vector3 = np.array([np.nan,7,6,np.nan,7,np.nan,9,np.nan,1,np.nan])
df = pd.DataFrame({"V1":vector1,"V2":vector2,"V3":vector3})
df
```

#### Â» Drop the rows if contains any NaN value (not permanent)

```python
df.dropna()
```

#### Â» Drop rows if every column has NaN values (not permanent)

```python
df.dropna(how="all") 
```

#### Â» Drop columns(variable) if contains any NaN value (not permanent)

```python
df.dropna(axis=1)
```

```python
df["null"] = np.nan
df
```

#### Â» Drop columns(variable) if every row has NaN values (permanent beacuse inplace is used)

```python
df.dropna(axis=1, how="all", inplace=True)
df
```
