#### ¬ª Create a dataframe containing NaN values 

```python
import numpy as np
import pandas as pd
vector1 = np.array([5,3,6,np.nan,7,2,8,np.nan,4,np.nan])
vector2 = np.array([10,1,5,3,2,8,np.nan,6,np.nan,np.nan])
vector3 = np.array([np.nan,7,6,np.nan,7,np.nan,9,np.nan,1,np.nan])
df = pd.DataFrame({"V1":vector1,"V2":vector2,"V3":vector3})
df
```

#### ¬ª Import "msno" to visualize NaN values and their ratio

```python
import missingno as msno
msno.bar(df)
```

The right bar of the graph shows values, while the left bar shows ratios. The bars represent actual values, while the gaps represent nan values. 

```python
msno.matrix(df)
```

The curve on the right shows the number of columns that do not contain NaN values in that row.

```python
import seaborn as sns
planets = sns.load_dataset("planets")
planets.head()
```

```python
planets.isnull().sum()
```

```python
msno.matrix(planets)
```

Before correcting NaN values, we should check if there is a relationship between them to violate any error in the dataset.

```python
msno.heatmap(planets)
```

We can use the heatmap to check whether mass and distance are correlated. In principle, there should not be a strong correlation, since distance from Earth is independent of a planet‚Äôs mass. However, missing values (NaNs) can distort correlation coefficients, so any observed relationship might be an artifact of incomplete data rather than a true dependency.
