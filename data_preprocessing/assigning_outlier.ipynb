```python
import numpy as np
import pandas as pd
vector1 = np.array([5,3,6,np.nan,7,2,8,np.nan,4,np.nan])
vector2 = np.array([10,1,5,3,2,8,np.nan,6,np.nan,11])
vector3 = np.array([np.nan,7,6,np.nan,7,np.nan,9,np.nan,1,np.nan])
df = pd.DataFrame({"V1":vector1,"V2":vector2,"V3":vector3})
df
```

```python
#assigning in numerical variables
```

```python
df["V1"].fillna(0)
```

```python
df["V1"].fillna(df["V1"].mean())
```

```python
#First way for all variables
```

```python
df.apply(lambda x: x.fillna(x.mean()), axis=0)
```

```python
#Second way
```

```python
df.mean()[:]
```

```python
df.fillna(df.mean()[:])
```

```python
df.fillna(df.mean()["V1":"V2"])
```

```python
df["V3"].fillna(df["V3"].median())
```

```python
#Third way
```

```python
df.where(pd.notna(df),df.mean(),axis="columns")
```
