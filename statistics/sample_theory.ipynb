```python
import numpy as np
```

```python
population = np.random.randint(0,80,100000) #the ages of a population
```

```python
population[0:10]
```

```python
#a sample
np.random.seed(115) #it provides the unchancing random values for every call
sample = np.random.choice(a=population,size=100)
sample[0:10]
```

```python
sample.mean()
```

```python
population.mean()
```

```python
np.random.seed(10)
s = []
for i in range(10):
    s.append(np.random.choice(a=population, size=100))
s
```

```python
s[0].mean()
```

```python
s[2].mean()
```

```python
sum_of_mean=0
for i in s:
    sum_of_mean += i.mean()
sum_of_mean / 10
```
