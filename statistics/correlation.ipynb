```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips["total_bill"] -= tips["tip"]
```

```python
tips.head()
```

```python
tips.plot.scatter("tip","total_bill")
```

```python
#Normality Assumption
from scipy.stats import shapiro
p1 = shapiro(tips["tip"])
print("p = {:.5f}".format(p1[1]))
```

```python
p2 = shapiro(tips["tip"])
print("p = {:.5f}".format(p2[1]))
```

```python
#Correlation Coefficient
tips["tip"].corr(tips["total_bill"]) #pearson correlation coefficient
```

```python
tips["tip"].corr(tips["total_bill"], method="spearman")
```

```python
#correlation significance test
from scipy.stats import pearsonr
```

```python
pval = pearsonr(tips["tip"], tips["total_bill"])
```

```python
print("Correlation Coefficient = {:.4f} , p = {:.5f}".format(pval[0],pval[1]))
```

```python
#Nonparametric Correlation Test
from scipy.stats import spearmanr
```

```python
pv = spearmanr(tips["tip"], tips["total_bill"])
```

```python
print("Correlation Coefficient = {:.4f} , p = {:.5f}".format(pv[0],pv[1]))
```
