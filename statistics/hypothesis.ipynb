```python
#shapiro-wilks Test
```

```python
from scipy.stats import shapiro
```

```python
shapiro(results) #(statistic of test, p value)
```

```python
#hypothesis test
```

```python
stats.ttest_1samp(results, popmean=170)
```

```python
# p-value is greater than 0.05, so H0 cannot be rejected.
# Since data is normally distributed (Shapiro-Wilk p > 0.05), t-test is appropriate.
# If the normality assumption was violated, we could use a non-parametric test such as the Wilcoxon signed-rank test instead.
```

```python
#non-parametric test
from statsmodels.stats.descriptivestats import sign_test
```

```python
sign_test(results, 170)
```
