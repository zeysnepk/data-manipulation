A confidence interval is a range of values,calculated from sample data that is likely to contain the true population parameter with a specific probability (e.g., 95% or 99%)

$\bar{X} \; \pm \; Z \cdot \frac{\sigma}{\sqrt{n}}$

- $\bar{X}$ = sample mean  
- $Z$ = Z-score for the desired confidence level (e.g., 1.96 for 95%)  
- $\sigma$ = population standard deviation  
- $n$ = sample size

---

#### ¬ª Import Numpy and create random integer array of 1000  to simulate prices ranges 10 and 110 given 1000 people

```python
import numpy as np
prices = np.random.randint(10,110,1000)
prices
```

#### ¬ª Calculate the mean of the prices array

```python
prices.mean()
```

#### ¬ª Calculate the %95 reliability price range by importing "statsmodels"

```python
import statsmodels.stats.api as sms
sms.DescrStatsW(prices).tconfint_mean()
```

The result shows that if the price is betweeen at 58.7 and 62.23, people are 95% likely to buy it.
