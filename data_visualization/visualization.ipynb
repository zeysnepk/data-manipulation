```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
```

```python
dias = sns.load_dataset("diamonds")
```

```python
df = dias.copy()
```

```python
df.head()
```

```python
cat_cut = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
```

```python
df.cut = df.cut.astype(CategoricalDtype(categories=cat_cut, ordered=True))
```

```python
df["cut"].head(1)
```

```python
df["cut"].value_counts()
```

```python
df["cut"].value_counts().plot.barh();
```

```python
df["cut"].value_counts().plot.barh().set_title("Class Frequencies of 'Cut 'Variable");
```

```python
(df["cut"]
 .value_counts()
 .plot.barh()
 .set_title("Class Frequencies of 'Cut 'Variable"));
```

```python
sns.barplot(x="cut", y=df.cut.index, data=df);
```
