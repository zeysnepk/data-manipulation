# üê± CatPlot
##### - A catplot (categorical plot) is a high-level plotting function in seaborn 
##### - Allows easy visualization of categorical data across different types of plots.
##### - It can create strip plots, swarm plots, boxplots, violin plots, bar plots, and count plots depending on the 'kind' parameter (e.g., kind="bar", kind="box", kind="violin").
##### - It is useful for visualizing relationships between a categorical variable and one or more numeric variables, and for comparing across multiple subgroups.
---

#### ¬ª Import Pandas for data manipulation and create dataframe for 'diamonds' dataset

```python
import seaborn as sns
dias = sns.load_dataset("diamonds")
df = dias.copy()
df.dtypes
```

#### ¬ª Order categories using CategoricalDtype

```python
from pandas.api.types import CategoricalDtype
cat = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories=cat, ordered=True))
df.head()
```

#### ¬ª This catplot displays the relationship between diamond cut categories and their prices, showing distributions per group using a default strip plot.

```python
sns.catplot(x="cut", y="price", data=df)
```

#### ¬ª This catplot shows the distribution of diamond prices across cut categories using violin plots, combining boxplot statistics with a kernel density estimate for each group.

```python
sns.catplot(x="cut", y="price", data=df,kind="violin")
```

#### ¬ª This catplot visualizes the distribution of diamond prices across cut categories using boxplots, highlighting medians, quartiles, and potential outliers for each group.

```python
sns.catplot(x="cut", y="price", data=df,kind="box")
```

#### ¬ª This catplot displays the average diamond price for each cut category using bar plots, with error bars indicating the standard deviation.

```python
sns.catplot(x="cut", y="price", data=df,kind="bar")
```

#### ¬ª This catplot creates a point plot showing the average diamond price for each cut category, separated by color groups with different markers, and includes vertical lines indicating standard deviation

```python
sns.catplot(x="cut",y="price",hue="color",kind="point",data=df)
```
