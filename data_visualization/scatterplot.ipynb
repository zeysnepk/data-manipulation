#### Â» Loading and previewing the 'tips' dataset

```python
import seaborn as sns
df = sns.load_dataset("tips").copy()
df.head()
```

#### Â» Creating a basic scatter plot of total bill vs tip

```python
sns.scatterplot(x=df.total_bill, y=df.tip)
```

#### Â» Scatter plot of total bill vs tip, colored by time of day

```python
sns.scatterplot(x=df.total_bill, y=df.tip, hue=df.time)
```

#### Â» Scatter plot of total bill vs tip, using color and marker style to show time of day

```python
sns.scatterplot(x=df.total_bill, y=df.tip, hue=df.time, style=df.time)
```

#### Â» Scatter plot of total bill vs tip, with color by day and marker style by time

```python
sns.scatterplot(x=df.total_bill, y=df.tip, hue=df.day, style=df.time)
```

#### Â» Scatter plot using color for time of day and marker style for day of week

```python
sns.scatterplot(x=df.total_bill, y=df.tip, hue=df.time, style=df.day)
```

#### Â» Scatter plot with marker size representing party size

```python
sns.scatterplot(x=df.total_bill, y=df.tip, size=df["size"])
```

#### Â» Scatter plot where both size and color represent party size

```python
sns.scatterplot(x=df.total_bill, y=df.tip, size=df["size"], hue=df["size"])
```
