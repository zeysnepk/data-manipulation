# ðŸ“Š Barplot
##### - A barplot is a type of chart that represents data using rectangular bars.
##### - The length or height of each bar corresponds to the value it represents.
##### - In seaborn, barplot is typically used to show the average (mean) of a numerical variable 
##### - It is useful for comparing group means and identifying patterns or differences between categories.
---

#### Â» Load and preview the 'diamonds' dataset

```python
import seaborn as sns
dias = sns.load_dataset("diamonds")
df = dias.copy()
df.head()
```

#### Â» Define the desired order of diamond 'cut' categories from lowest to highest quality

```python
cat_cut = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
```

#### Â» Convert the 'cut' column to an ordered categorical variable

```python
from pandas.api.types import CategoricalDtype
df.cut = df.cut.astype(CategoricalDtype(categories=cat_cut, ordered=True))
```

```python
df["cut"].head(1)
```

```python
df["cut"].value_counts()
```

#### Â» Plot the counts as a horizontal bar chart for better readability of labels

```python
df["cut"].value_counts().plot.barh();
```

#### Â» Add a descriptive title to the chart for clarity

```python
df["cut"].value_counts().plot.barh().set_title("Class Frequencies of 'Cut 'Variable");
```

#### Â» This version simply uses method chaining for cleaner and more readable code

```python
(df["cut"]
 .value_counts()
 .plot.barh()
 .set_title("Class Frequencies of 'Cut 'Variable"));
```

#### Â» Plot the counts as a vertical bar chart using the seaborn library

```python
sns.barplot(x="cut", y=df.cut.index, data=df);
```

#### Â» 'hue' adds coloring

```python
sns.barplot(x="cut", y=df.cut.index, data=df, hue="cut");
```

#### Â» This barplot compares the 'cut' categories against 'price',separated by diamond color where bar heights represent mean values and the black lines show standard deviation.

```python
sns.barplot(x="cut", y=df.price, data=df, hue="color");
```

#### Â» Calculate the mean diamond price for each combination of cut and color categories to observe the graph above

```python
df.groupby(["cut","color"],observed=False)["price"].mean()
```
