```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
```

```python
dias = sns.load_dataset("diamonds")
```

```python
df = dias.copy()
```

```python
cat = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
```

```python
df.cut = df.cut.astype(CategoricalDtype(categories=cat, ordered=True))
```

```python
df.head()
```

```python
sns.catplot(x="cut", y="price", data=df)
```

```python
sns.barplot(x="cut",y="price",hue="color",data=df) #mean price values
```

```python
df.groupby(["cut","color"])["price"].mean()
```
