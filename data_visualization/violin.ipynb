# ðŸŽ» Violin
##### - A violin plot combines a boxplot with a kernel density estimate, showing both summary statistics (median, quartiles, and outliers) and the full distribution shape of the data across categories.    
##### - It is useful for comparing distributions, detecting skewness, and identifying variability between groups.
---

#### Â» Load and preview the 'tips' dataset

```python
import seaborn as sns
df = sns.load_dataset("tips").copy()
df.head()
```

#### Â» This basic violin plot shows distribution of total bill

```python
sns.catplot(y=df.total_bill, kind="violin")
```

#### Â» This violin plot shows total bill across days of the week

```python
sns.catplot(y=df.total_bill, kind="violin", x=df.day)
```

#### Â» This violin plot shows total bill by day, split by sex using color

```python
sns.catplot(y=df.total_bill, kind="violin", x=df.day, hue=df.sex)
```
