#### Â» Importing Seaborn and loading the fMRI dataset

```python
import seaborn as sns
df = sns.load_dataset("fmri").copy()
df.head()
```

#### Â» Checking data types of each column

```python
df.dtypes
```

#### Â» Checking the shape of the dataset (rows, columns)

```python
df.shape
```

#### Â» Descriptive statistics of the 'timepoint' column

```python
df.timepoint.describe().T
```

#### Â» Descriptive statistics of the 'signal' column

```python
df.signal.describe().T
```

#### Â» Counting number of signal values for each timepoint

```python
df.groupby("timepoint")["signal"].count()
```

#### Â» Grouping by unique signal values and counting occurrences

```python
df.groupby("signal").count()
```

#### Â» Descriptive statistics of the signal grouped by timepoint

```python
df.groupby("timepoint")["signal"].describe()
```

#### Â» Checking unique subjects in the dataset

```python
df.subject.unique()
```

#### Â» Plotting signal over time

```python
sns.lineplot(x="timepoint",y="signal",data=df)
```

#### Â» Plotting signal over time by event type

```python
sns.lineplot(x="timepoint",y="signal",data=df, hue="event")
```

#### Â» Plotting signal over time by event type with different line styles

```python
sns.lineplot(x="timepoint",y="signal",data=df, hue="event", style="event")
```

#### Â» Plotting signal over time by event type with markers

```python
sns.lineplot(x="timepoint",y="signal",data=df, hue="event", style="event",markers=True)
```

#### Â» Plotting signal over time with markers and solid lines (no dashes)

```python
sns.lineplot(x="timepoint",y="signal",data=df, hue="event", style="event",markers=True,dashes=False)
```

#### Â» Plotting signal over time by brain region and event type using color and line style

```python
sns.lineplot(x="timepoint",y="signal",data=df, hue="region", style="event")
```
