##### - A heatmap is a data visualization technique that represents values in a matrix as colors.
##### - The color intensity corresponds to the magnitude of the values, making it easy to spot patterns, relationships, clusters, or anomalies at a glance.
##### - It is especially useful for analyzing correlations between variables or visualizing large numeric tables such as time series.
---

#### ¬ª Load and preview the 'flights' dataset

```python
import seaborn as sns
df = sns.load_dataset("flights").copy()
df.head()
```

```python
df.describe().T
```

```python
df.dtypes
```

#### ¬ª Pivot the DataFrame so that 'month' values become the index (y-axis) and 'year' values become the columns (x-axis), with 'passengers' as the cell values, creating the structured matrix format required for a heatmap

```python
df = df.pivot(index="month", columns="year", values="passengers")
```

```python
df
```

#### ¬ª This heatmap visualizes the pivoted DataFrame, where colors represent passenger counts across months (rows) and years (columns), making temporal patterns easy to detect

```python
sns.heatmap(df)
```

#### ¬ª This heatmap shows passenger counts across months and years with colors and annotates each cell with its exact integer value for clearer interpretation

```python
sns.heatmap(df, annot=True, fmt="d")
```

#### ¬ª This heatmap visualizes passenger counts across months and years with colors, annotates each cell with its integer value and adds grid lines (linewidth=.5) to clearly separate the cells

```python
sns.heatmap(df, annot=True, fmt="d", linewidth=.5)
```

#### ¬ª A heatmap without color bar

```python
sns.heatmap(df, annot=True, fmt="d", linewidth=.5, cbar=False)
```
