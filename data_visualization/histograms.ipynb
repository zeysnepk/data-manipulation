# ðŸ“‰ Histogram
##### - A histogram is used to display the distribution of a numerical variable by grouping values into bins.
##### - The x-axis shows intervals (bins) of the variable, while the y-axis shows the frequency of observations within each bin.
##### - It is useful for understanding the shape of the data distribution, detecting skewness, spread, and identifying possible outliers.
---

#### Â» Create dataframe for 'diamonds' dataset

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

#### Â» This histogram displays the distribution of diamond prices by grouping them into bins and showing their frequencies.  

```python
sns.histplot(df.price)
```

#### Â» This histogram shows the distribution of diamond prices with bins outlined as step-like lines instead of filled bars.  

```python
sns.histplot(df.price,element="step")
```

#### Â» This histogram visualizes the distribution of diamond prices using 1000 fine-grained bins with step-like outlines for detailed frequency patterns.

```python
sns.histplot(df.price,element="step",bins=1000)
```

#### Â» This histogram displays the distribution of diamond prices using 10 wide bins with step-like outlines, giving a coarse view of frequency patterns.

```python
sns.histplot(df.price,element="step",bins=10)
```

#### Â» This histogram shows the distribution of diamond prices with step-like outlines and overlays a KDE curve to estimate the probability density.  

```python
sns.histplot(df.price,element="step",kde=True)
```

#### Â» This KDE plot estimates and visualizes the probability density of diamond prices as a smooth continuous curve.  

```python
sns.kdeplot(df.price);
```

#### Â» This KDE plot shows the estimated probability density of diamond prices as a smooth curve with the area under the curve filled.  

```python
sns.kdeplot(df.price, fill=True)
```

#### Â» Use FacetGrid to overlay filled KDE curves of 'price' for each 'cut' level (hue), limit x to 0â€“10000, and add a legend
##### ! Crossing all categories ensures that every cut appears in the legend with consistent colors, making the plot useful for detecting differences in spread, skewness, and overall distribution patterns.

```python
(sns
 .FacetGrid(df,
               hue="cut",
               height=5,
               xlim=(0,10000))
 .map(sns.kdeplot, "price", fill=True)
 .add_legend());
```
