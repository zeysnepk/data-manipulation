```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("tips").copy()
df.head()
```

```python
sns.lmplot(x="total_bill", y="tip", data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", data=df, hue="smoker");
```

```python
sns.lmplot(x="total_bill", y="tip", data=df, hue="smoker", col="time");
```

```python
sns.lmplot(x="total_bill", y="tip", data=df, hue="smoker", col="time", row="sex");
```
